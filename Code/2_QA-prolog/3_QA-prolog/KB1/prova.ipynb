```python
import numpy as np
import os
```

```python
command_QAP = 'QA-Prolog --qmasm-args="--asd" --work-dir="/home/dadi/Thesis/Code/2_QA-prolog/3_QA-prolog/KB1/work"'
```

```python
command_qmasm = 'qmasm --format="numpy" -o="out.npz" --pin="Query.Valid := true" --solver="tabu" work/animal_unfold.qmasm'
```

```python
file = 'animal_unfold.pl'
```

```python
query = []
query.append("class(animal).")
query.append("class(X).")
query.append("subClass(seal, X).")
query.append("isA(reiny, X)")
query.append("isA(sardy, X)")
query.append("isA(Y, X)")
query.append("hasProperty(I1, ancestor, reiny_c).")
query.append("error(X, Y, Z)")
```

```python
n_qubit = []
for q in query:
    command = command_QAP + ' --query="' + q + '" ' + file
    os.system(command)
    os.system(command_qmasm)
    data = np.load("out.npz")
    print(len(data["syms"]))
    n_qubit.append(len(data["syms"]))
```

```python
for i in range(len(query)):
    print(query[i], n_qubit[i])
```
