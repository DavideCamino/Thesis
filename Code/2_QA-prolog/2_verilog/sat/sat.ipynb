```verilog
module sat (a, b, c, y);
   input a, b, c;
   output y;

   assign y = a & (b | c);
endmodule
```

`yosys sat.v synth.ys -b edif -o sat.edif`

`edif2qmasm -o="sat.qmasm" sat.edif`

`qmasm --solver="sim_anneal" --format="ocean" -o="ocean.py" --pin="sat.y := true" sat.qmasm`
